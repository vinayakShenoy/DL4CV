```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid_activation(x):
    return 1.0/(1+np.exp(-x))
```

```python
def sigmoid_deriv(x):
    return x*(1-x)
```

```python
def predict(X,W):
    preds = sigmoid_activation(X.dot(W))
    preds[preds<=0.5]=0
    preds[preds>0.5]=1
    return preds
```

```python
args = {
    "epochs":100,
    "alpha":0.01, #learning rate
    "batch-size":32
}
```

```python
(X,y) = make_blobs(n_samples=1000, n_features=2, centers=2,
                  cluster_std=1.5, random_state=1)
y = y.reshape((y.shape[0],1))

#insert a columns of 1s in last in feature matrix
X = np.c_[X, np.ones((X.shape[0]))]
(trainX, testX, trainY, testY) = train_test_split(X, y, test_size=0.5, 
                                                  random_state=42)
```

```python
print("[INFO] training...")
W = np.random.randn(X.shape[1],1)
losses = []
```

```python
for epoch in np.arange(0, args["epochs"]):
    preds = sigmoid_activation(trainX.dot(W))
    error = preds - trainY
    loss = np.sum(error**2)
    losses.append(loss)
    d = error*sigmoid_deriv(preds)
    gradient = trainX.T.dot(d)
    W += -args["alpha"]*gradient
    if epoch==0 or (epoch+1)%5==0:
        print("[INFO] epoch={}, loss={:7f}".format(int(epoch+1), loss))
```

```python
print("[INFO] evaluating")
preds = predict(testX, W)
print(classification_report(testY, preds))
```

```python
plt.style.use("ggplot")
plt.figure()
plt.title("Data")
plt.scatter(testX[:, 0], testX[:,1], marker="o", c=testY[:,0], s=30)

print(len(losses))

plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, args["epochs"]), losses)
plt.title("Training Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.show()
```

# Stochastic Gradient Descent

```python
print("[INFO] training...")
W = np.random.randn(X.shape[1],1)
losses = []
```

```python
def next_batch(X, y, batchSize):
    for i in np.arange(0, X.shape[0], batchSize):
        yield (X[i:i+batchSize], y[i:i+batchSize])
```

```python
for epoch in np.arange(0, args["epochs"]):
    epochLoss = []
    for (batchX, batchY) in next_batch(trainX, trainY, args["batch-size"]):
        preds = sigmoid_activation(batchX.dot(W))
        error = preds - batchY
        epochLoss.append(np.sum(error*2))
        d = error*sigmoid_deriv(preds)
        gradient = batchX.T.dot(d)
        W += -args["alpha"]*gradient
    loss = np.average(epochLoss)
    losses.append(loss)
    if epoch==0 or (epoch+1)%5==0:
        print("[INFO] epoch={}, loss={:.7f}".format(int(epoch+1),
                                                   loss))
```

```python
print("INFO evaluating")
preds = predict(testX, W)
print(classification_report(testY, preds))
```

```python
plt.style.use("ggplot")
plt.figure()
plt.title("Data")
plt.scatter(testX[:, 0], testX[:, 1], marker="o", c=testY[:, 0], s=30)

plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, args["epochs"]), losses)
plt.title("Training Loss")
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.show()
```

# Momemtum and Acceleration
- Momentum: Enables SGD to learn faster by focusing on dimensions whose gradient point in the same direction.
- Our previous weight rule
$$
W = W - \alpha\nabla_{W}f(W) 
$$
- With momentum V, scaled by $\gamma$
$$
V = \gamma V_{t-1} + \alpha\nabla_{W}f(W)
$$
$$
W = W - V_{t}
$$ 
- Nesterov acceleration can be conceptualized as a corrective update to the momentum which lets us obtain an approximate idea of where our parameters will be after the update

# Regularization
- L1 and L2 regularization(commonly known as weight decay), ElasticNet, are used to update the loss function itself, adding an additional parameter to constrain the capacity of the model.
- Helps us control our model capacity, ensuring that our models are better at making classifications on data points that they were not trained on, which is called generalization.
- Too much generalization is bad, causing underfitting, in which case our model performs badly on training data and is not able to model the relationship between the input data and output class labels.

## Updating our loss and weight update to include regularization
- In order to generalize and avoid underfitting, incorporate a penalty, called **regularization penalty** into the loss function.
- L2 regularization:
$$
R(W) = \sum_{i}\sum_{j}W^{2}_{i,j}
$$
- The larger tthe weight value is, the more influence it has on output prediction. So dimensions with larger weight values can almost singlehandedly control output prediction of classifier, which leads to overfitting.
- To mitigate the effect various dimensions have on our output classification, we apply regularization, thereby seeking $W$ values that take into account all dimensions rather than the few with large values. 
- Hurts training acc. but increases testing accuracy.
- To original loss function.
$$
L = \frac{1}{N}\sum^{N}_{i=1}L_{i}+\gamma R(W)
$$
- The second term is the regularization penalty. $\gamma$ variable is a hyperparameter that controls the amount or strength of the regularization we are applying.
- Expanded loss function
$$
L= \frac{1}{N}\sum^{N}_{i=1}[-log(\frac{e^{s_{y_{i}}}}{\sum_{j}e^{s_{y_{j}}}})] + \gamma \sum_{i}\sum_{j} W_{i,j}^2
$$
- Weight update expression
$$
W = W - \alpha\nabla_{W}f(W)-\gamma R(W)
$$

## Types of regularization
- L2 regularization, weight decay
$$
R(W) = \sum_{i}\sum_{j} W_{i,j}^2
$$
- L1 regularization
$$
R(W) = \sum_{i}\sum_{j} |W_{i,j}|
$$
- Elasticnet combines both
$$
R(W) = \sum_{i}\sum_{j} (\beta W_{i,j}^2+|W_{i,j}|)
$$

```python
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from imutils import paths
import import_ipynb
from utils import SimplePreprocessor, SimpleDatasetLoader
```

```python
args = {
    "dataset":"../dogs-v-cats/animals/animals/"
}
print("[INFO] LOADING IMAGES")
imagePaths = list(paths.list_images(args["dataset"]))
```

```python
sp = SimplePreprocessor(32,32)
sdl = SimpleDatasetLoader(preprocessors=[sp])
(data, labels) = sdl.load(imagePaths, verbose=500)
data = data.reshape((data.shape[0], 3072))
```

```python
le = LabelEncoder()
labels = le.fit_transform(labels)
(trainX, testX, trainY, testY) = train_test_split(data, labels, 
                                                 test_size=0.25, random_state=42)
```

```python
for r in (None, "l1", "l2"):
    print("[INFO] training model with {} penalty".format(r))
    model = SGDClassifier(loss="log", penalty=r, max_iter=500,
                         learning_rate="constant", tol=1e-3, eta0=0.01, random_state=42)
    model.fit(trainX, trainY)
    acc = model.score(testX, testY)
    print("[INFO] {} penalty accuracy: {:.2f}%".format(r, acc*100))
```
